```python
import numpy as np
from matplotlib.path import Path
import random

A=np.array([0,0])
B=np.array([1,0])
C=np.array([0.5,np.sqrt(3)/2])

def create_triangle(A, B, C):
    # Crear un objeto Path que representa el triángulo
    vertices = np.array([A, B, C, A])  # Los vértices del triángulo (cierre con A)
    return Path(vertices)

triangulo = create_triangle(A, B, C)

x=[]
y=[]

def extremo_random():
    return random.choice([A,B,C])

while True:
    p_random = np.random.rand(2)
    if triangulo.contains_point(p_random):
        break 

mitad = (extremo_random() - p_random) / 2
x.append(mitad[0])
y.append(mitad[1])

for i in range(1000):
    mitad= (extremo_random() - mitad) /2
    x.append(mitad[0])
    y.append(mitad[1])
```

```python
import matplotlib.pyplot as plt

plt.scatter(x, y, color='b', label='y vs x')  # Línea con puntos marcados
plt.xlabel('Eje X')  # Etiqueta del eje X
plt.ylabel('Eje Y')  # Etiqueta del eje Y
plt.title('Sierpinski')  # Título del gráfico
plt.show()
```

```python
import numpy as np
import random
from matplotlib.path import Path

# Definición de los extremos del triángulo
A = np.array([0, 0])
B = np.array([1, 0])
C = np.array([0.5,np.sqrt(3)/2])

# Crear el triángulo usando matplotlib.path.Path
triangle = Path([A, B, C, A])  # Cerramos el triángulo regresando al primer punto

# Función para seleccionar un extremo aleatorio
def extremo_random():
    return random.choice([A, B, C])

# Inicializar listas para acumular resultados
x = []
y = []

# Generar un punto aleatorio que esté dentro del triángulo
while True:
    p_random = np.random.rand(2)
    if triangle.contains_point(p_random):  # Verificar si está dentro del triángulo
        break  # Salir del bucle si está dentro

# Calcular la primera mitad
mitad = (extremo_random() + p_random) / 2
x.append(mitad[0])
y.append(mitad[1])

# Iterar para generar los demás puntos
for i in range(1000):
    mitad = (extremo_random() + mitad) / 2
    x.append(mitad[0])
    y.append(mitad[1])

import matplotlib.pyplot as plt

plt.scatter(x, y, color='b', label='y vs x')  # Línea con puntos marcados
plt.xlabel('Eje X')  # Etiqueta del eje X
plt.ylabel('Eje Y')  # Etiqueta del eje Y
plt.title('Sierpinski')  # Título del gráfico
plt.show()
```
